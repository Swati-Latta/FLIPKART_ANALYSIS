```python
f = open('Laptops_info.csv','wb')
f.write('Product_Name,Stars,Ratings,Reviews,CurrentPrice,MRP,Processor,RAM,Storage,ImageURL\n'.encode())




# Finding product name

for i in container:
    product= i.find('div',{'class':'_4rR01T'})
    Product_Name=product.text.split('-')[0].strip()
    print(Product_Name)
    
    
    
    # finding stars
    
    star=i.find('div',{'class':'_3LWZlK'})
    try:
        Stars= star.text
    except:
        Stars=0
        
        
     # finding rating and reviews
   
    Rating= i.find('span',{'class':'_2_R_DZ'})
    try:
        rat_rev = re.findall('\d+,?\d*',Rating.text)
        Ratings=rat_rev[0].replace(',','')
        Reviews=rat_rev[1].replace(',','')
    except:
        Ratings=0
        Reviews=0   
        
        
        
        # Finding current_price
     
    CurrentPrice= i.find('div',{'class':'_30jeq3 _1_WHN1'}).text.replace(',','')
    
    
    
    # finding MRP


    mrp = i.find('div',{'class':'_3I9_wc _27UcVY'})
    try:
        MRP=mrp.text.replace(',','')
    except: 
        MRP=0
        
        
        
        
    # finding product information

    info= i.findAll('li',{'class':'rgWa7D'})
    Processor = info[0].text
    RAM = info[1].text
    Storage = info[3].text
    
    
    
    # finding image URL

    Image = i.img
    ImageURL=Image.get('src')
    
    
    
    print(Product_Name,Stars, Ratings, Reviews, CurrentPrice, MRP, Processor ,RAM,Storage, ImageURL)
    f.write(f"{Product_Name},{Stars},{Ratings},{Reviews},{CurrentPrice},{MRP},{Processor},{RAM},{Storage},{ImageURL}\n".encode())
    print('\n\n')
f.close()            
```
