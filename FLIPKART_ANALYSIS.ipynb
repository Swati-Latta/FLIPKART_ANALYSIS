# Outliers

```python
for i in num:
    plt.figure()
    sns.boxplot(data=df,y=i)
```

```python
for c in num:
    q1=df[c].quantile(0.25)
    q3=df[c].quantile(0.75)
    iqr = q3-q1
    ul= q3+1.5* iqr
    ll= q3-1.5*iqr
    
    for i in df[c]:
        if i>ul:
            df[c]=df[c].replace(i,ul)
        elif i< ll:
             df[c]=df[c].replace(i,ll)
    
```

```python
for i in num:
    plt.figure()
    sns.boxplot(data=df,y=i)
```
